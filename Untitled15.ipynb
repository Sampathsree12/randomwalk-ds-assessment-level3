```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
books_tags=pd.read_csv(r"C:\Users\Naveen\OneDrive\Desktop\randomwalk-ds-assessment-level3\book_tags.csv")
books=pd.read_csv(r"C:\Users\Naveen\OneDrive\Desktop\randomwalk-ds-assessment-level3\books.csv")
ratings=pd.read_csv(r"C:\Users\Naveen\OneDrive\Desktop\randomwalk-ds-assessment-level3\ratings.csv")
```

```python
books_without_original_titles = books[books['title'].duplicated(keep=False) | books['title'].isnull()]
number_of_books_without_original_titles = len(books_without_original_titles)
print("Number of books without original titles:", number_of_books_without_original_titles)
```

```python
books= books.dropna(subset=['original_title'])
books_tags = books_tags[books_tags['goodreads_book_id'].isin(books['book_id'])]
ratings = ratings[ratings['book_id'].isin(books['book_id'])]
unique_books_count = books['book_id'].nunique()
print("Number of unique books after cleaning:", unique_books_count)
```

```python
unique_users = ratings['user_id'].nunique()
print(f'The number of unique users is: {unique_users}')
```

```python
unique_tags = books_tags['tag_id'].nunique()
print("Number of unique tags in book_tags.csv:", unique_tags)
```

```python
tag_counts = books_tags['tag_id'].value_counts().sort_values(ascending=False)
most_used_tag_id = tag_counts.idxmax()
print("Most frequently used tag_id:", most_used_tag_id)

```

```python
merged_df = pd.merge(books_tags, books[['book_id', 'title']], how='inner', left_on='goodreads_book_id', right_on='book_id')
title_tag_counts = merged_df.groupby('title')['tag_id'].count().sort_values(ascending=False)
most_tagged_book_title = title_tag_counts.idxmax()
print("Book with the most number of tag counts:", most_tagged_book_title)
```

```python

tag_counts = books_tags.groupby('tag_id')['goodreads_book_id'].count().reset_index(name='user_records')
tag_counts = tag_counts.sort_values(by='user_records', ascending=False)
top_20_tags = tag_counts.head(20)

```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_20_tags['tag_id'], top_20_tags['user_records'], color='blue')
plt.xlabel('Tag ID')
plt.ylabel('Number of User Records')
plt.title('Top 20 Tags by User Records')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
